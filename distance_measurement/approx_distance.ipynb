# object distance estimations

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
def calculate_distance(focal_length, original_radius, image_radius):
    return (focal_length * original_radius) / image_radius

img = cv2.imread('img7.jpeg')


gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
haar_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces_rect = haar_cascade.detectMultiScale(gray_img, 1.1, 9)
img_h,img_w,c=img.shape
for (x, y, w, h) in faces_rect:
    if (w>100):
      cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
      distance=calculate_distance(95,160*3.77,w)                                                                         #replace middle value with your face length in cm
      print("width and height of face:",w,h)
      print("x,y coordinate of image",x,y)
      cv2.putText(img, 'original:'+str(145), (50, 52), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)                 #just put original image distance(in cm) here
      cv2.putText(img, 'calculate:'+str(distance/3.77), (img_w-300, 52), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)


#img=cv2.resize(img,(250,180))
print("distance of man from camera is:",distance/3.77,"cm")
print("The shape of the image is",img_h,img_w)
cv2_imshow(img)
```
